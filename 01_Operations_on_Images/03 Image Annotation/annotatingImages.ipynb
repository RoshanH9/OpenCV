```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
image = cv2.imread("images/Apollo_11_Launch..webp", 1)
```

```python
plt.imshow(image[:, :, ::-1])
```

```python
# Drawing a Line on the image
imageLine = image.copy()
cv2.line(imageLine, (200,100), (400,100),(0, 0, 255), thickness= 5, lineType = cv2.LINE_4)
plt.imshow(imageLine[:, :, ::-1])
```

```python
imageCircle = image.copy()

cv2.circle(imageCircle, (450, 300), 100, (0, 0, 255), thickness= -2, lineType= cv2.LINE_AA)
plt.imshow(imageCircle[:, :, ::-1])
```

```python
# Drawing a rectangle
imageRectangle = image.copy()

cv2.rectangle(imageRectangle, (100, 200), (200, 300), (0, 0, 255), thickness=-2, lineType=cv2.LINE_AA)
plt.imshow(imageRectangle[:, :, ::-1])
```

```python
# Adding Text
imageText = image.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 0, 0)
fontThickness = 2
org = (150, 45) # bottom left origin for the text
cv2.putText(imageText, text, org, fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA)

plt.imshow(imageText[:, :, ::-1])
```
