### Splitting and merging

```python
newzealand_bgr = cv2.imread('images/New_Zealand_Lake.jpg', 1) # or cv2.IMREAD_COLOR
b, g, r = cv2.split(newzealand_bgr) # BGR openCV

plt.figure(figsize=[20, 5])
plt.subplot(1, 4, 1); plt.imshow(r, cmap = 'gray'); plt.title("Red Channel")
plt.subplot(1, 4, 2); plt.imshow(g, cmap = 'gray'); plt.title("Green Channel")
plt.subplot(1, 4, 3); plt.imshow(b, cmap = 'gray'); plt.title("Blue Channel")

final_img = cv2.merge((b, g, r))
plt.subplot(1, 4, 4); plt.imshow(final_img[:, : , ::-1]); plt.title("Merged Output")
```

```python
newzealand_rgb = cv2.cvtColor(newzealand_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(newzealand_rgb)
```

```python
newzealand_hls = cv2.cvtColor(newzealand_bgr, cv2.COLOR_BGR2HLS_FULL)
plt.imshow(newzealand_hls)
```

```python
cv2.imwrite('images/saved_img.png', newzealand_hls)
```
