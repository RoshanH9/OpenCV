### Cropping Images

```python
img_NZ_bgr = cv2.imread('images/New_Zealand_Lake.jpg', cv2.IMREAD_COLOR)
img_NZ_rgb = img_NZ_bgr[:, :, ::-1] # swapping color channels

plt.imshow(img_NZ_rgb)
```

```python
cropped = img_NZ_rgb[200:400, 300:600] # slicing
plt.imshow(cropped)
```

### Resizing Images

```python
resized_cropped_image_2x = cv2.resize(cropped, None, fx = 2, fy = 2)
plt.imshow(resized_cropped_image_2x)
```

```python
desired_width = 100
desired_height = 200
dim = (desired_width, desired_height)

resized = cv2.resize(cropped, dsize = dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized)
```

### Maintaining the Aspect Ratio

```python
desired_width = 100
aspect_ratio = desired_width / resized.shape[1]
desired_height = int(resized.shape[0] * aspect_ratio)
dim = (desired_width, desired_height)

resized_cropped = cv2.resize(resized, dsize = dim, interpolation = cv2.INTER_AREA)
plt.imshow(resized_cropped)
```

```python
img_lake_horz = cv2.flip(img_NZ_rgb, 1) # about y - axis
img_lake_vert = cv2.flip(img_NZ_rgb, 0) # about x - axis
img_lake_both = cv2.flip(img_NZ_rgb, -1) # about both axis

plt.figure(figsize = [18, 5])
plt.subplot(141);plt.imshow(img_lake_horz);plt.title("Horizontal Flip");
plt.subplot(142);plt.imshow(img_lake_vert);plt.title("Vertical Flip");
plt.subplot(143);plt.imshow(img_lake_both);plt.title("Both Flipped");
plt.subplot(144);plt.imshow(img_NZ_rgb);plt.title("Original");
```

```python
img_lake_horz_resized = cv2.resize(img_lake_horz, dsize = (1000,1000))
plt.subplot(1,2,1);plt.imshow(img_lake_horz_resized);plt.title("RESIZED")
plt.subplot(1,2,2);plt.imshow(img_lake_horz);plt.title("Original")
```
